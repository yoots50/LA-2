```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
import PIL
from PIL import Image
```

```python
IMAGE = "/content/drive/MyDrive/Colab Notebooks/smung.png" #      
img = Image.open(IMAGE)

imMatrix = np.array(img.convert("L")) / 255.0

shape0, shape1 = np.shape(imMatrix) #imMatrix shape  shape0, shape1 
short = 0
if (shape0 > shape1): #  
  short = shape1 #shape1 
else: #  
  short = shape0 #shape0 
k = short // 30 #    shape 30   k   

U, S, V = npl.svd(imMatrix)
imtemp = lambda k: (U[:, :k] @ np.diag(S[:k]) @ V.T[:, :k].T) * 255
im = imtemp(k)

m, n = np.shape(imMatrix)
partial, total = (k * (m + n) + k, m * n)
partial, total, partial / total

plt.imshow(im, cmap = 'gray')
plt.show()
```
