```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
import PIL
from PIL import Image
```

```python
IMAGE = "/content/drive/MyDrive/Colab Notebooks/smung.png" # 이미지의 경로 또는 이름을 여기에 입력
img = Image.open(IMAGE)

imMatrix = np.array(img.convert("L")) / 255.0

shape0, shape1 = np.shape(imMatrix) #imMatrix의 shape를 각각 shape0, shape1에 저장
short = 0
if (shape0 > shape1): #세로로 긴 이미지라면
  short = shape1 #shape1이 작음
else: #가로로 긴 이미지라면
  short = shape0 #shape0이 작음
k = short // 30 # 둘 중의 작은 shape를 30으로 나눈 값을 k로 만들어 줌 

U, S, V = npl.svd(imMatrix)
imtemp = lambda k: (U[:, :k] @ np.diag(S[:k]) @ V.T[:, :k].T) * 255
im = imtemp(k)

m, n = np.shape(imMatrix)
partial, total = (k * (m + n) + k, m * n)
partial, total, partial / total

plt.imshow(im, cmap = 'gray')
plt.show()
```
